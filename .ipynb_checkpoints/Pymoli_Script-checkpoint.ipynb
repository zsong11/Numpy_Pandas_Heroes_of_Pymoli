```python
import pandas as pd
import numpy as np
```

```python
path = "Resources/purchase_data.csv"
df = pd.read_csv(path)

df.head()
```

```python
playerlist = df["SN"].unique()
player_count = len(playerlist)
pd.DataFrame({"Total Players": [player_count]})
```

```python
unique_item_count = len(df["Item ID" ].unique())
average_price = df["Price"].mean()
num_purchase = df["Purchase ID"].count()
revenue = df["Price"].sum()

purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": unique_item_count,
                                       "Average Price": average_price,
                                       "Number of Purchases": num_purchase,
                                       "Total Revenue": revenue}, index=[0])

purchasing_analysis_df
```

```python
gender = df["Gender"]
gender_count = df.groupby("Gender")["SN"].nunique()

gender_percentage = (gender_count / player_count) * 100
gender_percentage = gender_percentage.round(2)

gender_demographics_df = pd.DataFrame({"Total Count": gender_count,
                                       "Percent of Players": gender_percentage,})
gender_demographics_df
```

```python
purchase_count_by_gender = df["Gender"].value_counts()

average_purchase_price = df.groupby("Gender")["Price"].mean()
total_purchase_value = df.groupby("Gender")["Price"].sum()
avg_purchase_total_per_person = df.groupby("Gender")["Price"].mean()

purchasing_analysis = pd.DataFrame({"Purchase Count": purchase_count_by_gender,
                                       "Average Purcahse Price": average_purchase_price.round(2),
                                       "Total Purchase Value": total_purchase_value,
                                       "Average Purchase Total per Person": (total_purchase_value/gender_count).round(2)})

purchasing_analysis
```

```python
age_label = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
df["Age Demographics"] = pd.cut(df["Age"], age_bins, labels=age_label)

purchase_count_age = df.groupby("Age Demographics")["SN"].nunique()
avg_purchase_age = df.groupby("Age Demographics")["Price"].nunique().sum()
total_purchase_value_age = df.groupby("Age Demographics")["SN"].nunique().sum()
avg_per_person_age = df.groupby("Age Demographics")["Price"].mean()
age_count_g_unique = df.groupby("Age Demographics")["SN"].nunique()

age_demographics_df = pd.DataFrame({"Total Count": purchase_count_age,
                                    "Percentage of Players": (purchase_count_age/total_purchase_value_age) * 100})

age_demographics_df
```

```python
purchase_count = df.groupby("Age Demographics")["Purchase ID"].count()
purchase_count_unique = df.groupby("Age Demographics")["SN"].nunique()
average_purchase_price = df.groupby("Age Demographics")["Price"].mean()
total_purchase_value = df.groupby("Age Demographics")["Price"].sum()
avg_total_purchase_per_person = df.groupby("Age Demographics")["Price"].sum()/purchase_count_unique

purchasing_analysis_df = pd.DataFrame({"Purchase Count": purchase_count,
                                       "Average Purcahse Price": average_purchase_price.round(2),
                                       "Total Purchase Value": total_purchase_value,
                                       "Avg Total Purchase per Person": (total_purchase_value/purchase_count_unique).round(2)})

purchasing_analysis_df
```

```python
user_total = df.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")
user_average = df.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
user_count = df.groupby(["SN"]).count()["Price"].rename("Purchase Count")


user_data = pd.DataFrame({"Total Purchase Value": user_total, 
                          "Average Purchase Price": user_average, 
                          "Purchase Count": user_count})


user_data["Average Purchase Price"] = user_data["Average Purchase Price"].map("${:,.2f}".format)
user_data["Total Purchase Value"] = user_data["Total Purchase Value"].map("${:,.2f}".format)
user_data = user_data.loc[:,["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]


user_data.sort_values("Total Purchase Value", ascending=False).head(5)
```

```python
item_data = df.loc[:,["Item ID", "Item Name", "Price"]]

total_item_purchase = item_data.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")
average_item_purchase = item_data.groupby(["Item ID", "Item Name"]).mean()["Price"]
item_count = item_data.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")

item_data_pd = pd.DataFrame({"Total Purchase Value": total_item_purchase, "Item Price": average_item_purchase, "Purchase Count": item_count})
item_data_pd["Item Price"] = item_data_pd["Item Price"].map("${:,.2f}".format)
item_data_pd ["Purchase Count"] = item_data_pd["Purchase Count"].map("{:,}".format)
item_data_pd["Total Purchase Value"] = item_data_pd["Total Purchase Value"].map("${:,.2f}".format)
item_data_pd = item_data_pd.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]


item_data_pd.sort_values("Purchase Count", ascending=False).head(5)
```
